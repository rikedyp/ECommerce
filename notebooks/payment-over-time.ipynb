We will take some columns from the `orders` data set and use them to plot variables over time.

The variables we will plot are, per day:

- Number of purchases
- Number of order approvals
- Number of orders delivered
- Number of expected deliveries

We need to convert the time stamps provided into numbers that we can do computation with. We only need 1 day precision, so we will convert time stamps into Dyalog Date Numbers, rounded down.

```apl
Timestamp2day ← {1 ⎕DT 2⊃¨': -'∘⎕VFI¨⊆⍵}
```

```apl
cols ← 'order_purchase_timestamp' 'order_approved_at' 'order_delivered_customer_date' 'order_estimated_delivery_date'
(purchase approval delivered expected)←⌊↓⍉Timestamp2day orders.Column cols
```

```apl
(p_count a_count d_count e_count) ← {⍺,≢⍵}⌸¨ purchase approval delivered expected
4↑p_count
```
